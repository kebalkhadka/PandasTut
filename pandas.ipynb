## Group by 

```python
import pandas as pd

var = pd.DataFrame({"Name":["a","b","c","d","a","b","a","b","c","d"],"s1":[1,3,2,1,3,4,2,5,3,2],"s2":[3,4,2,1,3,2,4,3,2,5]})
var

var_new = var.groupby("Name")
var_new
for x,y in  var_new:
    print(x)
    print("---------")
    print(y)
```

```python
var_new.get_group('a')
var_new
```
